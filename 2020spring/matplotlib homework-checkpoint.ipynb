```python
#自己操作 生成随机数绘制10点到12点每一分钟的气温,注意matplotlib默认不支持中文字符
import random
x=range(1,121)
y=[random.randint(20,35) for i in range(1,121)]
fig=plt.figure(figsize=(30,15),dpi=200,facecolor='pink')
plt.plot(x,y)
#调整刻度为时间
_xtick_labels=['10:{}'.format(i) for i in range(60)]
_xtick_labels+=['11:{}'.format(i) for i in range(60)]
plt.xticks(list(x)[::3],_xtick_labels[::3],rotation=45)#取步长、数字和字符串一一对应，两个参数数据的长度必须一样
plt.grid(alpha=0.2,linestyle='--')#绘制网格
#添加描述信息
from pylab import *
mpl.rcParams['font.sans-serif']=['SimHei']#记住这是导入中文字符的方式！
plt.xlabel('时间(min)')
plt.ylabel('温度(℃)')
plt.title('10点到12点气温变化情况')
plt.savefig('temperature.png')
plt.show()
```

```python
## 分析近两个星期上证指数的走势
from matplotlib import pyplot as plt
x=range(1,16)
y=[3011.67,3071.68,3034.51,2943.29,2996.76,2968.52,2923.49,2887.43,2789.25,2779.64,2728.76,2702.13,2745.62,2660.17,2722.44]
plt.figure(figsize=(30,15),dpi=100,facecolor='orange')
plt.plot(x,y,label='上证指数',color='pink',linestyle='--',linewidth=5)#可自行百度颜色代码
plt.legend(loc='center right')#设置图例
plt.ylabel('点数')
plt.title('近两个星期上证指数的走势')
plt.savefig('stocktrend.png')
plt.show()
```

legend参数
loc:图例位置
fontsize：字体大小
frameon：是否显示图例边框
ncol：图例的列的数量，一般为1
title：添加标题
shadow：为图例边框添加阴影
markerfirst：True表示图例标签在句柄右侧，False反之
markerscale：图例标记为原图标记的多少倍大小

# matplotlib散点图的绘制

```python
#绘制北京2016年3，10月份每天白天的最高气温

from matplotlib import pyplot as plt
from pylab import *
mpl.rcParams['font.sans-serif']=['SimHei']
t3=[11,17,16,11,12,11,12,6,6,7,8,9,12,15,14,17,18,21,16,17,20,14,15,15,15,19,21,22,22,22,23]
t10=[26,26,28,19,21,17,16,19,18,20,20,19,22,23,17,20,21,20,22,15,11,15,5,13,17,10,11,13,12,13,6]
x3=range(1,32)
x10=range(51,82)
plt.figure(figsize=(30,15),dpi=100,facecolor='cyan')#画图前先调整图像
plt.scatter(x3,t3,label='3月',color='pink')
plt.scatter(x10,t10,label='10月')#绘制散点图的代码在前,调整坐标轴的代码在后
_x=list(x3)+list(x10)
_xtick_label=['3月{}日'.format(i) for i in x3]
_xtick_label+=['10月{}日'.format(i-50) for i in x10]
plt.xticks(_x,_xtick_label,rotation=90)
plt.title('北京市2016年3，10月日间最高气温变化图')
plt.xlabel('时间')
plt.ylabel('温度')
plt.legend(loc='best')
plt.show()
```
